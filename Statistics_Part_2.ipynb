Q.1 What is hypothesis testing in statistics?

Answer>> Hypothesis testing is a statistical technique used to evaluate a
         hypothesis about a population parameter based on a sample of data. It's a crucial concept in statistics, and here's a breakdown:

Key Components:

1. Null Hypothesis (H0): A statement of no effect or no difference, e.g., "The average height of adults is 175 cm."
2. Alternative Hypothesis (H1): A statement of an effect or difference, e.g., "The average height of adults is not 175 cm."
3. Test Statistic: A numerical value calculated from the sample data to determine how likely it is that the null hypothesis is true.
4. P-Value: The probability of observing the test statistic (or a more extreme value) assuming the null hypothesis is true.
5. Significance Level (α): A predetermined threshold for the p-value (usually 0.05) to decide whether to reject the null hypothesis.

Steps in Hypothesis Testing:

1. Formulate the null and alternative hypotheses.
2. Choose a significance level (α).
3. Collect sample data.
4. Calculate the test statistic.
5. Determine the p-value.
6. Compare the p-value to the significance level (α).
7. Make a decision:
    - If p-value ≤ α, reject the null hypothesis (H0).
    - If p-value > α, fail to reject the null hypothesis (H0).

Types of Errors:

1. Type I Error: Rejecting a true null hypothesis (H0).
2. Type II Error: Failing to reject a false null hypothesis (H0).

Hypothesis testing is a powerful tool for making informed decisions based on data. It's essential to understand the concepts and apply them correctly to avoid errors and misinterpretations.

Q2. What is the null hypothesis, and how does it differ from the alternative hypothesis?

Answer>> In statistical hypothesis testing, the null hypothesis (H0) and
         alternative hypothesis (H1) are two mutually exclusive statements:

Null Hypothesis (H0):

1. A statement of no effect, no difference, or no relationship.
2. Typically represents the status quo or a default assumption.
3. Often denoted as "H0: μ = μ0" (population parameter equals a specific value).

Alternative Hypothesis (H1):

1. A statement of an effect, difference, or relationship.
2. Represents the research hypothesis or the expected outcome.
3. Often denoted as "H1: μ ≠ μ0" (population parameter is not equal to a specific value).

Key differences:

1. Direction: H0 typically states "no effect" or "no difference," while H1 states the opposite.
2. Burden of proof: H0 is assumed true until proven otherwise, while H1 is the hypothesis that requires evidence to support it.
3. Test direction: H0 is tested against H1, and the test result determines whether to reject or fail to reject H0.

Example:

Suppose we want to investigate whether a new fertilizer increases crop yield.

H0: μ = 100 (The average crop yield is 100 units with the new fertilizer.)
H1: μ > 100 (The average crop yield is greater than 100 units with the new fertilizer.)

In this example, H0 states no effect (no increase in yield), while H1 states an effect (an increase in yield). The test result will determine whether to reject or fail to reject H0.

Choosing the Significance Level

Q4. What does a P-value represent in hypothesis testing?

Answer>>In hypothesis testing, the P-value (probability value) represents the probability of observing a result as extreme or more extreme than the one you obtained, assuming that the null hypothesis (H0) is true.

What does the P-value
1. Probability of chance occurrence: The P-value indicates the likelihood of obtaining your results (or more extreme results) by chance, assuming H0 is true.
2. Compatibility with the null hypothesis: A small P-value (typically ≤ 0.05) suggests that the observed data are unlikely under H0, leading to rejection of H0. A large P-value (> 0.05) indicates that the data are compatible with H0, and you fail to reject H0.

Interpreting P-values
1. P-value ≤ 0.05: Reject H0 (statistically significant result)
2. P-value > 0.05: Fail to reject H0 (not statistically significant)
3. P-value close to 0.05: Borderline significance; consider additional analyses or interpretations

Q6. What are Type 1 and Type 2 errors in hypothesis testing?

Answer>> Type 1 and Type 2 Errors in Hypothesis Testing:

Type 1 Error:
A Type 1 error occurs when:

- Null hypothesis (H0) is true: But you reject it.
- False positive: You conclude that there is a statistically significant effect or difference when there isn't one.

Type 2 Error:
A Type 2 error occurs when:

- Null hypothesis (H0) is false: But you fail to reject it.
- False negative: You conclude that there is no statistically significant effect or difference when there actually is one.

Key Differences:
- Type 1 error: Rejecting a true null hypothesis (H0).
- Type 2 error: Failing to reject a false null hypothesis (H0).

Consequences:
- Type 1 error: Can lead to unnecessary changes or interventions.
- Type 2 error: Can lead to missed opportunities or failure to address important issues.

Controlling Errors:
- α (significance level): Controls Type 1 error rate.
- β (power): Controls Type 2 error rate.
- Sample size: Increasing sample size can reduce both Type 1 and Type 2 error rates.

By understanding Type 1 and Type 2 errors, you can design and interpret hypothesis tests more effectively, minimizing the risk of incorrect conclusions.

Q8.  What is the Z-test, and when is it used in hypothesis testing?

Answer>>The Z-test:

Definition:
The Z-test is a statistical test used to determine whether a sample mean is significantly different from a known population mean.

When to Use:
The Z-test is used in hypothesis testing when:

1. Large sample size: The sample size is large enough (typically n > 30) to assume normality.
2. Known population standard deviation: The population standard deviation (σ) is known.
3. Testing a mean: The goal is to test a hypothesis about a population mean.

Assumptions:
1. Normality: The sample data should be normally distributed or approximately normal for large samples.
2. Independence: Observations should be independent of each other.
3. Known population standard deviation: The population standard deviation (σ) should be known.

Z-test Formula:
Z = (x̄ - μ) / (σ / √n)

Where:

- x̄ = sample mean
- μ = known population mean
- σ = known population standard deviation
- n = sample size

Interpretation:
- Z-score: Calculate the Z-score using the formula.
- P-value: Look up the Z-score in a standard normal distribution table or use software to find the P-value.
- Compare to α: Compare the P-value to the chosen significance level (α) to make a decision about the null hypothesis.

The Z-test is a powerful tool for hypothesis testing, especially when dealing with large samples and known population standard deviations.

How Margin of Error Affects Confidence Interval:

Example:

CI = 175.2 cm ± 2.5 cm = (172.7 cm, 177.7 cm)

This means that we are 95% confident that the true average height of adults in the population lies between 172.7 cm and 177.7 cm.

Q20. What is an ANOVA test, and what are its assumptions?

Answer>>ANOVA Test:

Definition:
Analysis of Variance (ANOVA) is a statistical test used to compare means of three or more groups to determine if at least one group mean is different.

Types of ANOVA:
1. One-way ANOVA: Compares means of three or more groups based on one independent variable.
2. Two-way ANOVA: Compares means of three or more groups based on two independent variables.
3. Repeated-measures ANOVA: Compares means of three or more groups where the same subjects are used for each treatment.

Assumptions:
1. Normality: The data should be normally distributed within each group.
2. Equal variances: The variances of the data should be equal across all groups (homogeneity of variance).
3. Independence: Observations should be independent of each other.

Steps to Perform ANOVA:
1. Formulate the null and alternative hypotheses: The null hypothesis states that all group means are equal, while the alternative hypothesis states that at least one group mean is different.
2. Calculate the F-statistic: The F-statistic is calculated by dividing the mean square between groups (MSB) by the mean square within groups (MSW).
3. Determine the degrees of freedom: The degrees of freedom between groups (dfB) and within groups (dfW) are calculated.
4. Look up the critical F-value: The critical F-value is looked up in an F-distribution table or calculated using software.
5. Make a decision: If the calculated F-statistic is greater than the critical F-value or the P-value is less than the chosen significance level (α), reject the null hypothesis.

Post-Hoc Tests:
If the null hypothesis is rejected, post-hoc tests such as:

1. Tukey's HSD: Compare all possible pairs of group means.
2. Scheffé test: Compare all possible pairs of group means.
3. Bonferroni correction: Adjust the significance level to account for multiple comparisons.

```python
#Q5 Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Function to calculate and visualize Type 1 and Type 2 errors
def visualize_errors(population_mean, population_std, sample_size, sample_mean, alpha=0.05):
    # Step 1: Calculate the critical Z-value for the two-tailed test
    z_critical = stats.norm.ppf(1 - alpha / 2)  # Critical Z-value for two-tailed test

    # Step 2: Calculate the standard error
    standard_error = population_std / np.sqrt(sample_size)

    # Step 3: Calculate the Z-score for the sample
    z_score = (sample_mean - population_mean) / standard_error

    # Step 4: Create the plot for visualizing Type 1 and Type 2 errors
    x = np.linspace(population_mean - 4 * population_std, population_mean + 4 * population_std, 1000)

    # The normal distribution under the null hypothesis (mean = population_mean)
    y_null = stats.norm.pdf(x, population_mean, population_std)

    # Assume the alternative hypothesis is true, shift the mean (e.g., mean = population_mean + 3)
    true_mean = population_mean + 3  # Assuming the alternative hypothesis mean is shifted
    y_alternate = stats.norm.pdf(x, true_mean, population_std)

    # Plot the null hypothesis and alternative hypothesis distributions
    plt.plot(x, y_null, label="Null Hypothesis Distribution", color="blue")
    plt.plot(x, y_alternate, label="Alternative Hypothesis Distribution", color="orange", linestyle="--")

    # Fill the rejection regions (Type 1 error regions) under the null hypothesis
    plt.fill_between(x, 0, y_null, where=(x <= -z_critical) | (x >= z_critical), color="red", alpha=0.3, label="Rejection Region (Type 1 Error)")

    # Fill the Type 2 error region (where we fail to reject when the alternative is true)
    plt.fill_between(x, 0, y_alternate, where=(x >= -z_critical) & (x <= z_critical), color="green", alpha=0.3, label="Type 2 Error Region")

    # Mark the critical Z-values (rejection boundaries)
    plt.axvline(x=population_mean - z_critical * standard_error, color="black", linestyle="--", label=f"Critical Z = {-z_critical:.2f}")
    plt.axvline(x=population_mean + z_critical * standard_error, color="black", linestyle="--", label=f"Critical Z = {z_critical:.2f}")

    # Mark the sample mean on the plot
    plt.axvline(x=sample_mean, color="green", linestyle=":", label=f"Sample Mean = {sample_mean}")

    # Adding titles and labels
    plt.title("Visualization of Type 1 and Type 2 Errors")
    plt.xlabel("Test Statistic (Z-score)")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)

    # Show the plot
    plt.show()

# Simulate some sample data
np.random.seed(42)  # For reproducibility
population_mean = 100  # Population mean (μ₀)
population_std = 15  # Population standard deviation (σ)
sample_size = 50  # Sample size

# Simulate a sample mean (you can adjust this to test different hypotheses)
sample_mean = 105  # A sample mean that is different from the population mean

# Visualize Type 1 and Type 2 errors
visualize_errors(population_mean, population_std, sample_size, sample_mean, alpha=0.05)
```

```python
#Q9.Write a Python function to calculate the confidence interval for a sample mean and explain its significanc.
import numpy as np
from scipy import stats

# Function to calculate the confidence interval for a sample mean
def confidence_interval(sample, confidence_level=0.95):
    # Calculate sample mean and standard deviation
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)  # Sample standard deviation (ddof=1)
    sample_size = len(sample)

    # Find the t-score (for small sample sizes, we use the t-distribution)
    alpha = 1 - confidence_level
    t_score = stats.t.ppf(1 - alpha / 2, df=sample_size - 1)

    # Calculate the margin of error
    margin_of_error = t_score * (sample_std / np.sqrt(sample_size))

    # Calculate the confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

# Example usage
np.random.seed(42)  # For reproducibility

# Simulating a sample data (e.g., test scores)
sample_data = np.random.normal(75, 10, 50)  # Mean = 75, Std dev = 10, Sample size = 50

# Calculate the 95% confidence interval for the sample mean
lower, upper = confidence_interval(sample_data, confidence_level=0.95)

# Display the result
print(f"The 95% confidence interval for the sample mean is: ({lower:.2f}, {upper:.2f})")
```

```python
#Q11.Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.
def bayesian_inference(prior, likelihood, false_positive, evidence):
    # Apply Bayes' Theorem: P(H|E) = (P(E|H) * P(H)) / P(E)
    # where P(E) is the total probability of the evidence (marginal likelihood)

    # Marginal likelihood P(E)
    p_e = (likelihood * prior) + (false_positive * (1 - prior))

    # Posterior P(H|E)
    posterior = (likelihood * prior) / p_e

    return posterior

# Given data for Bayesian inference
prior = 0.01  # Prior probability (P(H)): 1% of people have the disease
likelihood = 0.9  # Likelihood (P(E|H)): 90% sensitivity of the test
false_positive = 0.05  # False Positive Rate (P(E|~H)): 5% of people without the disease test positive
evidence = 1  # Evidence (Positive test result)

# Calculate the posterior probability
posterior = bayesian_inference(prior, likelihood, false_positive, evidence)

# Display the result
print(f"The probability that the patient has the disease given a positive test result is: {posterior:.4f}")
```

```python
#Q13.Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data.
import numpy as np

# Function to calculate expected frequencies for Chi-square test
def calculate_expected_frequencies(observed):
    # Calculate row sums, column sums, and total sum
    row_sums = np.sum(observed, axis=1)  # Sum of each row
    col_sums = np.sum(observed, axis=0)  # Sum of each column
    total_sum = np.sum(observed)         # Total sum of all observations

    # Initialize an empty array to store expected frequencies
    expected = np.zeros_like(observed, dtype=float)

    # Calculate expected frequencies for each cell
    for i in range(observed.shape[0]):  # Iterate over rows
        for j in range(observed.shape[1]):  # Iterate over columns
            expected[i, j] = (row_sums[i] * col_sums[j]) / total_sum

    return expected

# Example observed data (2x2 contingency table)
observed_data = np.array([[50, 30],  # 50 males like product, 30 males don't
                          [20, 40]]) # 20 females like product, 40 females don't

# Calculate the expected frequencies
expected_data = calculate_expected_frequencies(observed_data)

# Display the results
print("Observed Data (Contingency Table):")
print(observed_data)
print("\nExpected Frequencies:")
print(expected_data)
```

```python
#Q16.Implement an F-test using Python to compare the variances of two random samples.
import numpy as np
import scipy.stats as stats

# Function to perform F-test
def f_test(sample1, sample2):
    # Calculate the variances of both samples
    var1 = np.var(sample1, ddof=1)  # Sample variance for sample1
    var2 = np.var(sample2, ddof=1)  # Sample variance for sample2

    # F-statistic is the ratio of the larger variance to the smaller variance
    f_statistic = max(var1, var2) / min(var1, var2)

    # Degrees of freedom for the samples
    df1 = len(sample1) - 1  # Degrees of freedom for sample1
    df2 = len(sample2) - 1  # Degrees of freedom for sample2

    # Calculate the p-value for the F-statistic
    p_value = 1 - stats.f.cdf(f_statistic, df1, df2)  # Right-tail probability

    # Display the results
    print(f"F-statistic: {f_statistic}")
    print(f"P-value: {p_value}")

    # Decision rule: If p-value < 0.05, reject H0 (variances are different)
    if p_value < 0.05:
        print("Reject the null hypothesis: The variances are significantly different.")
    else:
        print("Fail to reject the null hypothesis: The variances are not significantly different.")

# Example: Simulate two random samples
np.random.seed(42)  # For reproducibility

# Sample 1 (Normal distribution with mean=10, std=2, size=30)
sample1 = np.random.normal(loc=10, scale=2, size=30)

# Sample 2 (Normal distribution with mean=10, std=4, size=30)
sample2 = np.random.normal(loc=10, scale=4, size=30)

# Perform F-test to compare the variances
f_test(sample1, sample2)
```

```python
#Q19.Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Function to check the assumptions for ANOVA
def check_anova_assumptions(groups):
    # Check normality of each group using the Shapiro-Wilk test
    print("Checking normality for each group using Shapiro-Wilk test:")
    for i, group in enumerate(groups):
        stat, p_value = stats.shapiro(group)
        print(f"Group {i+1}: Shapiro-Wilk test statistic = {stat:.4f}, p-value = {p_value:.4f}")
        if p_value < 0.05:
            print(f"Group {i+1} does not follow a normal distribution (reject H0).\n")
        else:
            print(f"Group {i+1} follows a normal distribution (fail to reject H0).\n")

    # Check for equal variances across groups using Levene's test
    print("Checking equal variances (homoscedasticity) using Levene's test:")
    stat, p_value = stats.levene(*groups)
    print(f"Levene's test statistic = {stat:.4f}, p-value = {p_value:.4f}")
    if p_value < 0.05:
        print("The variances are significantly different (reject H0 for equal variances).\n")
    else:
        print("The variances are equal (fail to reject H0 for equal variances).\n")

    # Visualize the distribution of data using boxplots
    plt.figure(figsize=(8, 6))
    plt.boxplot(groups, patch_artist=True, vert=False)
    plt.title('Boxplot of Groups to Visualize Variances')
    plt.xlabel('Values')
    plt.ylabel('Groups')
    plt.show()

# Example usage with simulated data
np.random.seed(42)  # For reproducibility

# Simulate data for four groups
group1 = np.random.normal(loc=10, scale=2, size=30)  # Group 1: mean=10, std=2
group2 = np.random.normal(loc=12, scale=2, size=30)  # Group 2: mean=12, std=2
group3 = np.random.normal(loc=14, scale=2, size=30)  # Group 3: mean=14, std=2
group4 = np.random.normal(loc=16, scale=2, size=30)  # Group 4: mean=16, std=2

# List of groups
groups = [group1, group2, group3, group4]

# Check assumptions for ANOVA
check_anova_assumptions(groups)
```

```python
#Q20.import numpy as np
import pandas as pd
import scipy.stats as stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
import seaborn as sns
import matplotlib.pyplot as plt

# Function to perform Two-Way ANOVA
def perform_two_way_anova(data, dependent_variable, factor1, factor2):
    # Perform Two-Way ANOVA using statsmodels
    model = ols(f'{dependent_variable} ~ C({factor1}) * C({factor2})', data=data).fit()
    anova_table = sm.stats.anova_lm(model, typ=2)

    # Display ANOVA table
    print("Two-Way ANOVA Results:")
    print(anova_table)

    # Interpretation of results
    if anova_table['PR(>F)'][0] < 0.05:
        print(f"Reject the null hypothesis for {factor1}: There is a significant effect of {factor1} on {dependent_variable}.")
    else:
        print(f"Fail to reject the null hypothesis for {factor1}: No significant effect of {factor1} on {dependent_variable}.")

    if anova_table['PR(>F)'][1] < 0.05:
        print(f"Reject the null hypothesis for {factor2}: There is a significant effect of {factor2} on {dependent_variable}.")
    else:
        print(f"Fail to reject the null hypothesis for {factor2}: No significant effect of {factor2} on {dependent_variable}.")

    if anova_table['PR(>F)'][2] < 0.05:
        print(f"Reject the null hypothesis for interaction: There is a significant interaction between {factor1} and {factor2}.")
    else:
        print(f"Fail to reject the null hypothesis for interaction: No significant interaction between {factor1} and {factor2}.")

# Example: Simulate data for Two-Way ANOVA
np.random.seed(42)

# Simulate data for two factors (factor1 and factor2) with different levels
factor1 = ['A', 'B']  # Factor 1 with two levels: 'A' and 'B'
factor2 = ['X', 'Y']  # Factor 2 with two levels: 'X' and 'Y'

# Create a dataframe with simulated dependent variable values
data = []
for f1 in factor1:
    for f2 in factor2:
        # Simulate 30 data points for each combination of factor levels
        values = np.random.normal(loc=10 if f1 == 'A' else 12, scale=2, size=30)
        for value in values:
            data.append([f1, f2, value])

# Create a pandas DataFrame
df = pd.DataFrame(data, columns=[f'Factor1', f'Factor2', 'DependentVariable'])

# Perform the two-way ANOVA
perform_two_way_anova(df, 'DependentVariable', 'Factor1', 'Factor2')

# Visualize the interaction effect using seaborn's interaction plot
plt.figure(figsize=(8, 6))
sns.pointplot(data=df, x='Factor1', y='DependentVariable', hue='Factor2', dodge=True, markers='o', linestyles='-', ci=None)
plt.title('Interaction Plot: Effect of Factor1 and Factor2 on Dependent Variable', fontsize=16)
plt.xlabel('Factor 1', fontsize=12)
plt.ylabel('Dependent Variable', fontsize=12)
plt.show()
```

```python
#Q21.import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Function to plot the F-distribution
def plot_f_distribution(df1, df2):
    # Define the range of x values (F-statistic values)
    x = np.linspace(0, 5, 1000)

    # Calculate the F-distribution's probability density function (PDF)
    y = stats.f.pdf(x, df1, df2)

    # Create the plot
    plt.figure(figsize=(8, 6))
    plt.plot(x, y, label=f'F-distribution (df1={df1}, df2={df2})', color='b')

    # Add labels and title
    plt.title(f'F-distribution with df1={df1} and df2={df2}', fontsize=16)
    plt.xlabel('F-statistic', fontsize=12)
    plt.ylabel('Probability Density', fontsize=12)

    # Show the plot
    plt.grid(True)
    plt.legend()
    plt.show()

# Example: Visualize the F-distribution for different degrees of freedom
df1_values = [5, 10]  # Degrees of freedom for the numerator (between-group variance)
df2_values = [10, 20]  # Degrees of freedom for the denominator (within-group variance)

# Plot for multiple (df1, df2) combinations
for df1 in df1_values:
    for df2 in df2_values:
        plot_f_distribution(df1, df2)
```

```python
#Q22.Perform a one-way ANOVA test in Python and visualize the results with boxplots to compare group mean.
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# Function to perform One-Way ANOVA and visualize the results
def perform_one_way_anova(data, dependent_variable, group_variable):
    # Perform the one-way ANOVA using scipy
    groups = [data[data[group_variable] == group][dependent_variable] for group in data[group_variable].unique()]
    f_statistic, p_value = stats.f_oneway(*groups)

    # Display the ANOVA results
    print(f"ANOVA Results: F-statistic = {f_statistic:.4f}, p-value = {p_value:.4f}")

    # Interpret the results
    if p_value < 0.05:
        print("Reject the null hypothesis: At least one group mean is significantly different.")
    else:
        print("Fail to reject the null hypothesis: No significant difference between group means.")

    # Visualize the data with a boxplot
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=group_variable, y=dependent_variable, data=data, palette='Set2')
    plt.title(f'Boxplot of {dependent_variable} by {group_variable}')
    plt.xlabel(group_variable)
    plt.ylabel(dependent_variable)
    plt.show()

# Simulate example data for three groups
np.random.seed(42)
group1 = np.random.normal(loc=10, scale=2, size=30)  # Group 1: mean=10, std=2
group2 = np.random.normal(loc=12, scale=2, size=30)  # Group 2: mean=12, std=2
group3 = np.random.normal(loc=14, scale=2, size=30)  # Group 3: mean=14, std=2

# Create a DataFrame
data = pd.DataFrame({
    'Group': ['Group1'] * 30 + ['Group2'] * 30 + ['Group3'] * 30,
    'Score': np.concatenate([group1, group2, group3])
})

# Perform the one-way ANOVA and visualize the results
perform_one_way_anova(data, 'Score', 'Group')
```

```python
#Q23.Simulate random data from a normal distribution, then perform hypothesis testing to evaluate the means.
import numpy as np
import scipy.stats as stats

# Step 1: Simulate data from a normal distribution
np.random.seed(42)  # For reproducibility
sample_size = 100  # Number of data points
population_mean = 50  # Hypothesized population mean
population_std = 10  # Standard deviation of the population

# Simulate the sample data
sample_data = np.random.normal(loc=population_mean, scale=population_std, size=sample_size)

# Step 2: State the Hypotheses
# H₀: The sample mean is equal to the hypothesized population mean (50)
# H₁: The sample mean is not equal to the hypothesized population mean (50)

# Step 3: Perform the one-sample t-test
t_statistic, p_value = stats.ttest_1samp(sample_data, population_mean)

# Step 4: Interpret the results
alpha = 0.05  # Significance level

print(f"T-statistic: {t_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between the sample mean and the population mean.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the sample mean and the population mean.")
```

```python
#Q24.Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.
import numpy as np
import scipy.stats as stats

# Step 1: Simulate data from a normal distribution
np.random.seed(42)  # For reproducibility
sample_size = 30  # Number of data points
population_variance = 25  # Hypothesized population variance
sample_data = np.random.normal(loc=0, scale=np.sqrt(population_variance), size=sample_size)

# Step 2: Calculate sample variance (s^2)
sample_variance = np.var(sample_data, ddof=1)  # ddof=1 for sample variance

# Step 3: State the hypotheses
# H₀: The population variance is equal to the hypothesized population variance (σ² = 25)
# H₁: The population variance is not equal to the hypothesized population variance (σ² ≠ 25)

# Step 4: Calculate the test statistic (Chi-square statistic)
hypothesized_variance = 25  # Hypothesized population variance (σ₀²)
chi_square_statistic = (sample_size - 1) * sample_variance / hypothesized_variance

# Step 5: Find the critical values from the Chi-square distribution
alpha = 0.05  # Significance level
df = sample_size - 1  # Degrees of freedom

# Find the critical values for the two-tailed test
critical_value_lower = stats.chi2.ppf(alpha / 2, df)
critical_value_upper = stats.chi2.ppf(1 - alpha / 2, df)

# Step 6: Interpret the results
print(f"Chi-square Statistic: {chi_square_statistic:.4f}")
print(f"Critical Values: Lower = {critical_value_lower:.4f}, Upper = {critical_value_upper:.4f}")

if chi_square_statistic < critical_value_lower or chi_square_statistic > critical_value_upper:
    print("Reject the null hypothesis: The population variance is significantly different from the hypothesized value.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the sample variance and the hypothesized population variance.")
```

```python
#Q25.import numpy as np
import scipy.stats as stats

# Function to perform Z-test for proportions
def z_test_proportions(x1, n1, x2, n2):
    # Step 1: Calculate sample proportions
    p1 = x1 / n1  # Proportion for group 1
    p2 = x2 / n2  # Proportion for group 2

    # Step 2: Calculate pooled proportion
    P = (x1 + x2) / (n1 + n2)

    # Step 3: Calculate the Z-statistic
    z_statistic = (p1 - p2) / np.sqrt(P * (1 - P) * (1 / n1 + 1 / n2))

    # Step 4: Calculate the p-value (two-tailed test)
    p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))  # two-tailed

    # Step 5: Output the results
    print(f"Z-statistic: {z_statistic:.4f}")
    print(f"P-value: {p_value:.4f}")

    # Step 6: Interpret the results (assuming alpha = 0.05)
    alpha = 0.05
    if p_value < alpha:
        print("Reject the null hypothesis: There is a significant difference between the proportions.")
    else:
        print("Fail to reject the null hypothesis: There is no significant difference between the proportions.")

# Example data
# Group 1: 80 successes out of 200 trials
# Group 2: 90 successes out of 220 trials
x1, n1 = 80, 200  # successes, trials for group 1
x2, n2 = 90, 220  # successes, trials for group 2

# Perform Z-test for proportions
z_test_proportions(x1, n1, x2, n2)
```

```python
#Q26.Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results.
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Function to perform the F-test for comparing variances
def f_test_variances(data1, data2, alpha=0.05):
    # Step 1: Calculate sample variances
    s1_squared = np.var(data1, ddof=1)  # Sample variance of data1
    s2_squared = np.var(data2, ddof=1)  # Sample variance of data2

    # Step 2: Calculate the F-statistic (larger variance over smaller variance)
    f_statistic = max(s1_squared, s2_squared) / min(s1_squared, s2_squared)

    # Step 3: Degrees of freedom for both samples
    df1 = len(data1) - 1  # Degrees of freedom for sample 1
    df2 = len(data2) - 1  # Degrees of freedom for sample 2

    # Step 4: Calculate the p-value using the F-distribution
    p_value = 2 * min(stats.f.cdf(f_statistic, df1, df2), 1 - stats.f.cdf(f_statistic, df1, df2))  # Two-tailed test

    # Step 5: Output the results
    print(f"F-statistic: {f_statistic:.4f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print("Reject the null hypothesis: The variances are significantly different.")
    else:
        print("Fail to reject the null hypothesis: There is no significant difference between the variances.")

    return f_statistic, p_value, df1, df2

# Example Data
# Group 1: 50 random data points from normal distribution with mean=0 and std=1
# Group 2: 50 random data points from normal distribution with mean=0 and std=2
np.random.seed(42)  # For reproducibility
data1 = np.random.normal(loc=0, scale=1, size=50)  # Group 1 data
data2 = np.random.normal(loc=0, scale=2, size=50)  # Group 2 data

# Perform F-test for variances
f_statistic, p_value, df1, df2 = f_test_variances(data1, data2)

# Visualize the distributions of the two datasets
plt.figure(figsize=(10, 6))
plt.hist(data1, bins=15, alpha=0.7, label='Group 1 (Variance = 1)', color='blue')
plt.hist(data2, bins=15, alpha=0.7, label='Group 2 (Variance = 4)', color='red')
plt.title("Comparison of Variances: Group 1 vs Group 2")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```
